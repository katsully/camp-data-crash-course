```python
%pylab inline
# this installs the pandas library and creates a shorthand for it
import pandas as pd
```

```python
# read all the complaints from the CSV file and store it in the complaints 
# variable
path = 'DOB_Complaints_Received.csv'
complaints = pd.read_csv(path)
```

```python
# see the first 10 complaints from the data set
complaints[0:10]
```

```python
# Disposition Code A3 is full stop work order
# This creates a sub data set of all complaints that resulted in a full
# stop work order
stop_work = complaints[[x for x in complaints['Disposition Code']=='A3']]
```

```python
# let's us see the subset (it will show us the first 5 and last 5)
stop_work
```

```python
# prints how many rows (ie how many complaints) in the dataset
len(stop_work)
```

```python
# gives us the five most common House Streets in the data set and
# how many times they occured
stop_work['House Street'].value_counts()[:5].sort_values(ascending=False)
```

```python
# shows us all possible values for the Status column in the dataset
complaints.Status.unique()
```

```python
# making a sub dataset of all active complaints
active = complaints[[x for x in complaints.Status == 'ACTIVE']]
```

```python
# see the first 10 active complaints
active[0:10]
```

```python
# Complaint Category 81 is an elevator accident
# This creates a sub data set from our active dataset of all active elevator accidents
active_elevator_accidents = active[[x for x in active['Complaint Category'] == '81']]
```

```python
# preview our new dataset
active_elevator_accidents
```

```python
# this creates a dataset of all elevator accidents both active and closed
all_elevator_accidents = complaints[[x for x in complaints['Complaint Category'] == '81']]
```

```python
# look at the first ten entries in the dataset
all_elevator_accidents[0:10]
```

```python
# this creates a list of strings containing all streets where an elevator 
# accident occured
addresses = all_elevator_accidents['House Street'].tolist()
```

```python
# my list of house streets
addresses
```

```python
# I want to get rid of all the funny spacing issues in the list of streets
new_addresses = [s.strip() for s in addresses]
```

```python
# new and improved list with less funny spacing
new_addresses
```

```python
# do the same thing for house numbers and zip codes
house_numbers = all_elevator_accidents['House Number'].tolist()
zip_codes = all_elevator_accidents['ZIP Code'].tolist()

new_house_numbers = [s.strip() if isinstance(s, str) else "" for s in house_numbers]
new_zip_codes = [s.strip() if isinstance(s, str) else "" for s in zip_codes]
```

```python
# print all the house numbers where an elevator accident occured
new_house_numbers
```

```python
# bad_places is a list of strings containing the house number, street, and
# zip code for an elevator accident occured
bad_places = [a + " " + b + " " + c for a,b,c in zip(new_house_numbers, new_addresses, new_zip_codes)]
```

```python
# a list of places not to take an elevator
bad_places
```

```python
# figure out where elevator accidents happened the most often
from collections import Counter 
most_common = Counter(bad_places).most_common(5)
```

```python
# print places where you really don't want to take an elevator
most_common
```

```python
# replace the Date Entered columns with datetime objects
# datetime objects are much easier for a computer to parse than strings
complaints['Date Entered'] = pd.to_datetime(complaints['Date Entered'])
```

```python
# look at the new Date Entered column
# looks similar to us, but for python this is much easier
complaints[0:10]
```

```python
# create a histogram plotting complaints by month
complaints['Date Entered'].dt.month.hist(bins=12, figsize=(10,6))
xlabel('Month')
ylabel('number of complaints')
title('distribution of complaints amongst month')
```

```python
# let's make a plot - that shows trends over time
%pylab inline

complaints.groupby(complaints['Date Entered']).size().plot(figsize=(20,6))
```

```python
# we can easily specify a smaller date range
complaints.groupby(complaints['Date Entered']).size()['2000-02-01':'2000-02-15'].plot(figsize=(20,6))
```

```python
# let's create a sub dataset of all complaints from this uptick in complaints
outliers = complaints[[x for x in complaints['Date Entered']=='2000-02-11']]
```

```python
# let's look at the outliers, look at the Complaint Category column
outliers
```

```python
# let's find the top 5 category complaints for that day
outliers['Complaint Category'].value_counts()[:5].sort_values(ascending=False)
```

```python
# let's look at all of the unique community boards
# this is a 3 digit number representing a certain area of brooklyn
complaints['Community Board'].unique()
```

```python
# how many unique community boards were entered in all of the complaints?
len(complaints['Community Board'].unique())
```

```python
# However, look at the data, some are strings, some are numbers, some are 
# single numbers (and they're suppose to be 3), some have funny characters.
# We want to remove all complaints with invalid community board numbers

# this code keeps all complaints were the community board number has 3 digits
# and no non-digit characters
cleaned_data = complaints[[len(str(x))==3 and str(x).isdigit() for x in complaints['Community Board']]]
```

```python
# now let's look at our 'cleaned' data
cleaned_data['Community Board'].unique()
```

```python
# now we have duplicates because some entries are strings and some are numbers
# let's convert everything to a number
cleaned_data[['Community Board']] = cleaned_data[['Community Board']].astype(int)
```

```python
# let's look again
cleaned_data['Community Board'].unique()
```

```python
# let's find the top 5 areas where complaints are made
cleaned_data['Community Board'].value_counts()[:5].sort_values(ascending=False)
```
